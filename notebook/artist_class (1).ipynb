```python
#Reading Files into one dataframe
import pandas as pd
import tqdm
import glob
counter=0
data=pd.DataFrame()
for filename in glob.glob('./data/*.csv'):
    if filename=='./all_data.csv':
        pass
    elif counter==0:
        data=pd.read_csv(filename,header=None)
    else:
        df=pd.read_csv(filename,header=None)
        data=data.append(df)
    counter=counter+1
    print(counter)
    print(filename)
```
